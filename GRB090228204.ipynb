```python
import warnings

warnings.simplefilter("ignore")
import numpy as np

np.seterr(all="ignore")
from astropy.io import fits
```

```python
import matplotlib.pyplot as plt
from pathlib import Path
from threeML import *
from threeML.io.package_data import get_path_of_data_file
```

## Examining the GBM catalog GRB090228204

```python
gbm_catalog = FermiGBMBurstCatalog()
gbm_catalog.query_sources("GRB090228204")
```

```python
ra = 106.800
dec = -24.300
trigger = "GRB090228204"
```

```python
grb_info = gbm_catalog.get_detector_information()[trigger]

gbm_detectors = grb_info["detectors"]
source_interval = grb_info["source"]["fluence"]
background_interval = grb_info["background"]["full"]
best_fit_model = grb_info["best fit model"]["fluence"]
model_catalog = gbm_catalog.get_model(best_fit_model, "fluence")[trigger]
```

## Downloading the data

```python
dload = download_GBM_trigger_data("bn090228204", detectors=gbm_detectors)
```

## Temporal binning as defined in Burgess et. al. 2019

```python
source = ['-0.009--0.002', '-0.002-0.019','0.019-0.034','0.034-0.081','0.081-0.101', '0.101-0.155','0.155-0.260','0.260-0.433']
```

## Measurement for time bin 1

```python
fluence_plugins = []
time_series = {}
for det in gbm_detectors:

    ts_cspec = TimeSeriesBuilder.from_gbm_cspec_or_ctime(
        det, cspec_or_ctime_file=dload[det]["cspec"], rsp_file= "output_bad/bn090228204_" + det + ".rsp2"#dload[det]["rsp"]
    )

    ts_cspec.set_background_interval(*background_interval.split(','))
    ts_cspec.save_background(f"{det}_bkg.h5", overwrite=True)

    ts_tte = TimeSeriesBuilder.from_gbm_tte(
        det,
        tte_file=dload[det]["tte"],
        rsp_file= "output_bad/bn090228204_" + det + ".rsp2" #dload[det]["rsp"],
        #restore_background=f"{det}_bkg.h5",
    )
    
    ts_tte.set_background_interval(*background_interval.split(','))

    time_series[det] = ts_tte

    ts_tte.set_active_time_interval(source[0])

    ts_tte.view_lightcurve(-10, 10)

    fluence_plugin = ts_tte.to_spectrumlike()

    if det.startswith("b"):

        fluence_plugin.set_active_measurements("250-30000")

    else:

        fluence_plugin.set_active_measurements("9-900")

    #fluence_plugin.rebin_on_background(1.0)

    fluence_plugins.append(fluence_plugin)
```

## Defining the model

```python
cpl = Cutoff_powerlaw(piv=200)
#band = Band(piv=200)
ptsrc = PointSource(trigger, ra, dec, spectral_shape=cpl)
model = Model(ptsrc)
```

```python
model.display(complete=True)
```

## Setting priors for the model

```python
model.GRB090228204.spectrum.main.Cutoff_powerlaw.K.bounds = (1e-10,1e10)

model.GRB090228204.spectrum.main.Cutoff_powerlaw.K.prior = Log_normal(mu = -2, sigma=5)
model.GRB090228204.spectrum.main.Cutoff_powerlaw.index.prior = Gaussian(mu=-1, sigma=0.5)
model.GRB090228204.spectrum.main.Cutoff_powerlaw.xc.prior = Truncated_gaussian(lower_bound=10, upper_bound=10000,mu=200, sigma=300)
```

```python
from threeML.bayesian.sampler_base import SamplerBase
```

## Cloning the model and setting up the bayesian analysis class

```python
new_model = clone_model(model)

bayes = BayesianAnalysis(new_model, DataList(*fluence_plugins))

# share spectrum gives a linear speed up when
# spectrumlike plugins have the same RSP input energies
bayes.set_sampler("multinest", share_spectrum=True)
```

## Running the sampler

```python
bayes.sampler.setup(n_live_points=600)
bayes.sample()
```

```python
bayes.restore_median_fit()
bayes.results.display()
```

## Examining the fitted model

```python
fig = display_spectrum_model_counts(bayes, min_rate=10)
fig.axes[0].set_ylim(1e-9)
```

## Spectral parameters from Burgess et. al. 2019

```python
cpl = Cutoff_powerlaw(piv=200, xc = 339, index = -0.77)
ptsrc = PointSource(trigger, ra, dec, spectral_shape=cpl)
model_burgess = Model(ptsrc)
#model_burgess.display(complete=True)
```
